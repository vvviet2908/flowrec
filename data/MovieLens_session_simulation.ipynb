```python
import datetime as dt
import pandas as pd
import numpy as np
import time

def preprocess():
    rnames = ['userId', 'movieId', 'rating','timestamp']
    ratings_df = pd.read_table('https://raw.githubusercontent.com/vvviet2908/Recommender_System/main/ratings.dat', sep='::',
              header=None, names=rnames, engine='python')
    # ALL TRAIN SET (positive-only)
    all_train_data = ratings_df[ratings_df['rating']==5]
    all_train_data = ratings_df.drop(['rating'], axis=1)

    # only users 1 < rated_movies < 101
    all_train_data = all_train_data.groupby("userId").filter(lambda x: len(x) > 1 and len(x) < 1000)
    all_train_data.columns = ['SessionId', 'ItemId', 'Timestamps']
    all_train_data= all_train_data[['SessionId', 'Timestamps', 'ItemId']]
    all_train_data.to_csv('Movielens1M_positive.csv', index=False)
    return all_train_data
preprocess()
```
